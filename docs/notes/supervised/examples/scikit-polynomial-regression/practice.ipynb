```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

import pandas as pd
```

```python
df = pd.read_csv('data.csv')
X = df[['Var_X']]
y = df[['Var_Y']]
```

```python
# Create polynomial features (prediction plane)
poly_feat = PolynomialFeatures(degree=1)
X_poly = poly_feat.fit_transform(X)

print(X_poly)
```

```python
# Make and fit the polynomial regression model
poly_model = LinearRegression(fit_intercept=False).\
             fit(X_poly,y)
```

```python
# plot model line.
# 1. Sort X values in order to run them through the poly_model.predict()
# 2. poly_model.predict() will output predictions in sorted order
# 3. Compare with scatter plot of original data. scatter will sort data

x_sorted = X.values.copy()
x_sorted.sort(axis=0)
plt.plot(x_sorted,
         poly_model.predict(poly_feat.fit_transform(x_sorted)))

plt.scatter(X, y)
```

```python
y_model = poly_model.predict(X_poly)
y_model.sort(axis=0)
plt.plot(x_sorted,y_model)
plt.scatter(X, y)
```
